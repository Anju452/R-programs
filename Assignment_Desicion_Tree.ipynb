```python
# Outlier Detection: User defined Function to calculate Upper Extreme and Lower Extreme value
def outlier_detection(data,colname):
  q1 = data[colname].quantile(0.25)
  q3 = data[colname].quantile(0.75)
  iqr = q3 - q1

  upper_extreme = q3 + (1.5 * iqr)
  lower_extreme = q1 - (1.5 * iqr)

  return lower_extreme, upper_extreme,q1,q3
```

### **1. What are some common hyperparameters of decision tree models, and how do they affect the model's performance?**

Decision tree models have several important hyperparameters that influence their performance, including model complexity, accuracy, and risk of overfitting. Some common hyperparameters are:

1. Max Depth (max_depth)
Description: Limits the depth (number of levels) of the tree.
Effect:
A deeper tree can capture more complexity and patterns in the data but can also overfit to noise.
A shallower tree generalizes better but may underfit, missing important details in the data.
2. Minimum Samples Split (min_samples_split)
Description: Specifies the minimum number of samples required to split a node.
Effect:
Higher values prevent the tree from splitting nodes with very few samples, reducing overfitting.
Lower values allow more splits, potentially capturing more detailed relationships but increasing overfitting risks.
3. Minimum Samples Leaf (min_samples_leaf)
Description: The minimum number of samples required to be at a leaf node.
Effect:
A higher value encourages larger leaf nodes, which smooths the model and reduces overfitting.
A smaller value can result in more complex models but may lead to overfitting on small sample sizes.
4. Max Features (max_features)
Description: Limits the number of features considered when looking for the best split.
Effect:
Reducing the number of features can help reduce overfitting and speed up training.
Increasing it allows the model to consider more features, which may improve accuracy but also risks overfitting.
5. Criterion (criterion)
Description: The function to measure the quality of a split. Common options are "gini" (Gini impurity) and "entropy" (information gain).
Effect:
Gini tends to produce larger and simpler trees, favoring faster splits.
Entropy can produce more balanced splits but can be computationally slower.
6. Max Leaf Nodes (max_leaf_nodes)
Description: Limits the number of leaf nodes in the tree.
Effect:
A smaller number of leaf nodes results in a simpler tree and may prevent overfitting.
A larger number increases the tree's complexity, possibly leading to overfitting if not properly tuned.
7. Minimum Impurity Decrease (min_impurity_decrease)
Description: A node will only be split if this split induces a decrease in impurity greater than or equal to this value.
Effect:
It prevents unnecessary splits by requiring a minimum reduction in impurity.
Higher values can lead to a more generalized model (less splits), while lower values allow for finer distinctions in the data.
8. Splitter (splitter)
Description: Determines the strategy used to choose the split at each node. Options include "best" and "random."
Effect:
Best uses the best possible split based on the criterion, which can result in higher accuracy but may be slower.
Random introduces randomness into the splitting process, which can reduce overfitting in certain datasets.
9. Class Weight (class_weight)
Description: Assigns weights to the classes in classification problems to handle imbalanced datasets.
Effect:
Balanced weights help the model focus on underrepresented classes, improving its ability to handle imbalanced datasets.

### **2. What is the difference between the Label encoding and One-hot encoding?**

Label Encoding and One-Hot Encoding are two common techniques used to convert categorical data into numerical form so that machine learning models can work with it. Here's how they differ:

1. Label Encoding
Description: Each unique category in the categorical data is assigned a unique integer label. For example, if you have a column with three categories: ["Red", "Green", "Blue"], Label Encoding might map them to [0, 1, 2].
2. One-Hot Encoding
Description: Each unique category is represented as a binary vector. For n unique categories, you get n new binary columns. A value of 1 indicates the presence of the category, while 0 indicates absence.

Conclusion:
Use Label Encoding when there is an inherent ordinal relationship among the categories.
Use One-Hot Encoding for nominal categories to avoid introducing artificial relationships into the data.
