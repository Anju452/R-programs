```python
number_list = list(range(1, 7589))
```

```python
exrate['t'] = number_list
```

```python
# Create a new column 'Squared' that is the square of 'Numbers'
exrate['Squared'] = exrate['t'] ** 2

print("\nDataFrame with Squared Column:")
print(exrate)
```

```python
print(exrate)
```

```python
train = exrate.head(7000)
test = exrate.tail(588)
```

```python
train
```

```python
test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Linear Model
import numpy as np
import statsmodels.formula.api as smf

linear_model = smf.ols('Ex_rate~t',data=train).fit() #Footfalls - target col, t - independent variable
pred_linear = pd.Series(linear_model.predict(pd.DataFrame(test['t']))) # get prediction for test data
rmse_linear = np.sqrt(mean_squared_error(test['Ex_rate'],pred_linear)) # there is no direct built in function for rmst so take sqrt of MSE
rmse_linear # RMSE should be minimum. try all models. Whichever model will give you min, RMSE that will be our final model
```

```python
# Quadratic Model

quad = smf.ols('Ex_rate~t+Squared',data=train).fit()
pred_quad = pd.Series(quad.predict(pd.DataFrame(test[['t','Squared']])))
rmse_quad = np.sqrt(mean_squared_error(test['Ex_rate'],pred_quad))
rmse_quad
```

```python
# Additive Seasonality Model - Pass from Jan to Nov

add_sea = smf.ols('Ex_rate~	month',data=train).fit()
pred_add_sea = pd.Series(add_sea.predict(pd.DataFrame(test[['month']])))
rmse_add_sea = np.sqrt(mean_squared_error(test['Ex_rate'],pred_add_sea))
rmse_add_sea
```
