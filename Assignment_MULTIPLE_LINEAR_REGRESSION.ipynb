```python
# Outlier Detection: User defined Function to calculate Upper Extreme and Lower Extreme value
def outlier_detection(data,colname):
  q1 = data[colname].quantile(0.25)
  q3 = data[colname].quantile(0.75)
  iqr = q3 - q1

  upper_extreme = q3 + (1.5 * iqr)
  lower_extreme = q1 - (1.5 * iqr)

  return lower_extreme, upper_extreme,q1,q3
```

# **What is Normalization & Standardization and how is it helpful?**

Normalization and standardization are two techniques commonly used in data preprocessing to adjust the values of numerical data features so they can be more effectively used in machine learning algorithms. Both methods aim to bring the data within a specific range or distribution, making the algorithms more efficient and improving the performance of models.

### **Normalization**
Normalization is the process of scaling the data into a specific range, usually between 0 and 1. The formula for normalization (min-max scaling) is:

\[
X_{norm} = \frac{X - X_{min}}{X_{max} - X_{min}}
\]

Where:
- \( X \) is the original data.
- \( X_{min} \) is the minimum value of the feature.
- \( X_{max} \) is the maximum value of the feature.
  
**When to use Normalization:**
- When the data does not follow a Gaussian distribution (i.e., it is not normally distributed).
- When using machine learning algorithms that do not assume any distribution of data, such as k-nearest neighbors (KNN) or neural networks.
- When dealing with features that have different ranges, to bring them to a common scale.

### **Standardization**
Standardization transforms the data to have a mean of zero and a standard deviation of one. The formula for standardization (z-score normalization) is:

\[
X_{std} = \frac{X - \mu}{\sigma}
\]

Where:
- \( X \) is the original data.
- \( \mu \) is the mean of the feature.
- \( \sigma \) is the standard deviation of the feature.
  
**When to use Standardization:**
- When the data follows a Gaussian distribution or approximately does so.
- When using machine learning algorithms that assume the data is normally distributed, such as linear regression, logistic regression, or principal component analysis (PCA).
- When the scale of features varies but the data's overall distribution is of interest.

### **Why Normalization & Standardization Are Helpful**
1. **Improving Model Performance:** Some machine learning algorithms, especially those involving distance metrics (like KNN) or gradient descent (like neural networks), perform better when features are on a similar scale. This prevents features with larger ranges from dominating the algorithm.
   
2. **Faster Convergence:** In optimization problems, especially those using gradient-based methods, having features on a similar scale allows the algorithm to converge more quickly, leading to faster training times.

3. **Handling Outliers:** Standardization can help mitigate the impact of outliers, as it centers the data and reduces the influence of extreme values.

4. **Improving Accuracy:** By ensuring that features contribute equally to the result, these techniques help models generalize better to new data, improving their accuracy.

In summary, normalization and standardization are essential preprocessing steps that can significantly impact the effectiveness of a machine learning model by ensuring that the data is properly scaled and centered.

# **What techniques can be used to address multicollinearity in multiple linear regression?**

Multicollinearity occurs when two or more independent variables in a multiple linear regression model are highly correlated, meaning they provide redundant information. This can lead to problems like unstable estimates of regression coefficients, inflated standard errors, and reduced model interpretability. Addressing multicollinearity is crucial for improving the reliability of a regression model. Here are some techniques to handle it:

### 1. **Remove Highly Correlated Predictors**
   - **Correlation Matrix:** Calculate the correlation matrix to identify pairs of variables that have high correlation coefficients (typically above 0.7 or 0.8). You can then remove one of the correlated variables.
   - **Variance Inflation Factor (VIF):** Compute the VIF for each predictor. A VIF value greater than 10 (or sometimes even 5) indicates high multicollinearity. Consider removing predictors with high VIF values.

### 2. **Principal Component Analysis (PCA)**
   - PCA is a dimensionality reduction technique that transforms correlated variables into a smaller set of uncorrelated components (principal components). These components can then be used as predictors in the regression model, effectively eliminating multicollinearity.

### 3. **Regularization Techniques**
   - **Ridge Regression:** Ridge regression adds a penalty to the regression model proportional to the sum of the squared coefficients (L2 regularization). This shrinks the coefficients of less important variables, reducing multicollinearity.
   - **Lasso Regression:** Lasso regression adds a penalty based on the absolute value of the coefficients (L1 regularization). This can shrink some coefficients to zero, effectively performing variable selection and reducing multicollinearity.

### 4. **Combine or Aggregate Correlated Variables**
   - Sometimes, you can combine correlated variables into a single predictor. For example, you might calculate the average or sum of highly correlated variables to create a new feature.

### 5. **Centering the Data**
   - Centering involves subtracting the mean of each variable from its values, which can help reduce multicollinearity by making the variables less correlated. This is particularly useful when dealing with interaction terms.

### 6. **Partial Least Squares Regression (PLS)**
   - PLS is another dimensionality reduction technique like PCA, but it takes the dependent variable into account when constructing the components. It finds the directions that maximize the covariance between the predictors and the response variable, reducing multicollinearity.

### 7. **Remove or Combine Data**
   - If you have a small dataset, consider collecting more data, as multicollinearity is often exacerbated in small samples. Alternatively, you can aggregate or filter the data to reduce multicollinearity.

### 8. **Use of Domain Knowledge**
   - Sometimes, domain knowledge can guide you in deciding which variables are redundant and can be excluded from the model.

### 9. **Ridge or Lasso Regularization**
   - Using regularization techniques like Ridge or Lasso can help mitigate the effects of multicollinearity by shrinking or eliminating coefficients of less important predictors.

### 10. **Cross-Validation**
   - Employ cross-validation to check how the model performs with different subsets of data. This can help ensure that the multicollinearity is not affecting the model's generalization capability.

By applying these techniques, you can mitigate the effects of multicollinearity, leading to more stable and interpretable regression models.
